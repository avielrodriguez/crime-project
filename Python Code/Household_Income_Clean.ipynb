```python
import pandas as pd
import re
import os
import pymysql.cursors


# CSV cleanup magic begins...
def read_csv_with_header_detection(csv_file):
    # Read the CSV file
    df = pd.read_csv(csv_file)

    # Extract the household income ranges from the DataFrame's column headers
    headers = df.columns[1:].tolist()

    # Dictionary to store data for each subheading (as separate DataFrames)
    subheading_tables = {}
    current_subheading = None

    for index, row in df.iterrows():
        name = row[df.columns[0]].strip()

        if name.isupper():
            current_subheading = name
            subheading_tables[current_subheading] = pd.DataFrame()  # Create an empty DataFrame for the subheading
        else:
            if current_subheading not in subheading_tables:
                subheading_tables[current_subheading] = pd.DataFrame()  # Create an empty DataFrame if not exists

            # Convert the original headers to household income ranges
            income_ranges = headers
            income_values = row[1:].tolist()

            # Create a DataFrame with the current entry and concatenate it to the subheading's DataFrame
            entry = {'Name': name, **dict(zip(income_ranges, income_values))}
            entry_df = pd.DataFrame(entry, index=[0])
            subheading_tables[current_subheading] = pd.concat([subheading_tables[current_subheading], entry_df], ignore_index=True)

    return subheading_tables


# Helper function to escape special characters in a value
def escape_value(value):
    if isinstance(value, str):
        return value.replace("'", "''")
    return value

# Updated logic to remove $, commas, and spaces and replace with underscores
def sanitize_column_name(column_name):
    return re.sub(r'[\$, ]', '_', column_name).lower()

# Check if the table exists in the database    
def table_exists(cursor, table_name):
    query = f"SHOW TABLES LIKE '{table_name}';"
    cursor.execute(query)
    return len(cursor.fetchall()) > 0

### Making changes into the actual database via connection... ###
def create_tables(subheading_tables, connection):
    try:
        with connection.cursor() as cursor:
            for subheading, table in subheading_tables.items():
                table_name = re.sub(r'\W+', '_', subheading).lower()
                table_columns = [f"{sanitize_column_name(col)} VARCHAR(255)" for col in table.columns]
                create_table_query = f"CREATE TABLE IF NOT EXISTS {table_name} ({', '.join(table_columns)});"
                cursor.execute(create_table_query)
        connection.commit()
        print("Tables created successfully!")
    except Exception as e:
        print("Error:", e)
###...###

def generate_sql_create_table_statements(subheading_tables):
    sql_statements = []

    for subheading, table in subheading_tables.items():
        table_name = re.sub(r'\W+', '_', subheading).lower()
        # Update the column names to the new format and enclose in backticks
        table_columns = [f"`{sanitize_column_name(col)}` VARCHAR(255)" for col in table.columns]
        columns = ', '.join(table_columns)
        sql_statement = f"CREATE TABLE IF NOT EXISTS {table_name} ({columns});"
        sql_statements.append(sql_statement)

    return sql_statements

# Read the database password from the text file
def get_database_password():
    password_file = "db_password.txt"
    with open(password_file, 'r') as file:
        password = file.readline().strip()
    return password

# Establish connection...
def get_connection():
    host = 'localhost'
    user = 'avielr'
    password = get_database_password()  # Get the password from the text file
    database = 'crime_data_analysis'

    return pymysql.connect(host=host, user=user, password=password, database=database)

def generate_sql_insert_statements(subheading_tables):
    sql_statements = []

    for subheading, table in subheading_tables.items():
        table_name = re.sub(r'\W+', '_', subheading).lower()
        # Update the column names to the new format and enclose in backticks
        table_columns = [f"`{sanitize_column_name(col)}`" for col in table.columns]

        # Get the primary key column name (assuming it's 'id')
        primary_key_column = 'id'

        for _, row in table.iterrows():
            # Check if the primary key exists in the table
            if primary_key_column not in table.columns:
                columns = ', '.join(table_columns)
                values = ', '.join([f"'{escape_value(val)}'" for val in row.values])
                sql_statement = f"INSERT INTO {table_name} ({columns}) VALUES ({values});"
                sql_statements.append(sql_statement)
            else:
                # Check if the primary key value already exists in the table
                primary_key_value = row[primary_key_column]
                check_query = f"SELECT COUNT(*) FROM {table_name} WHERE {primary_key_column} = '{primary_key_value}';"

                # Execute the query to check if the data already exists
                with get_database_password() as conn:
                    with conn.cursor() as cursor:
                        cursor.execute(check_query)
                        count = cursor.fetchone()[0]
                        if count == 0:  # Data does not exist, add the INSERT statement
                            columns = ', '.join(table_columns)
                            values = ', '.join([f"'{escape_value(val)}'" for val in row.values])
                            sql_statement = f"INSERT INTO {table_name} ({columns}) VALUES ({values});"
                            sql_statements.append(sql_statement)

    return sql_statements


def execute_sql_statements(sql_statements, connection):
    try:
        with connection.cursor() as cursor:
            for sql_statement in sql_statements:
                cursor.execute(sql_statement)
        connection.commit()
        print("Data inserted successfully!")
    except Exception as e:
        print("Error:", e)


def main():
    # Fetch the CSV file...
    csv_file = "C:/Users/humbl/Desktop/MySQL Project Workspace/data/household_income_data.csv"
    subheading_tables = read_csv_with_header_detection(csv_file)

    # Get the MySQL connection
    connection = get_connection()

    # Create tables in MySQL server
    create_tables(subheading_tables, connection)

    # Generate SQL INSERT statements
    sql_statements = generate_sql_insert_statements(subheading_tables)

    # Execute the SQL INSERT statements in MySQL server
    execute_sql_statements(sql_statements, connection)

    # Close the connection after executing all queries
    connection.close()

if __name__ == "__main__":
    main()
```
